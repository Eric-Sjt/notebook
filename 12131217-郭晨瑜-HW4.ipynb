```python
import numpy as np
def diffnew(x, y, M=1):
    tem_x = x.pop(M-1)
    x.insert(0, tem_x)
    tem_y = y.pop(M-1)
    y.insert(0, tem_y)
    A = y
    N = len(x)
    for j in range(1, N):
        for k in range(N-1, j-1, -1):
            A[k] = (A[k] - A[k-1]) / (x[k] - x[k-j])
    x0 = x[0]
    df = A[1]
    prod = 1
    for k in range(1, len(A)-1):
        prod = prod * (x0 - x[k])
        df = df + prod * A[k+1]
    return A, df
```

## <span style="color:red">Solution:

We can know the Simpson's rule :
$$\int_{x_0}^{x_2} {f(x)} \,{\rm d}\approx \frac{h}{3}(f_0+4f_1 +f_2)\tag{1}$$
For $f(x) = x^2$,
$$\int_{a}^{b} {f(x)} \,{\rm d}\approx \frac{h}{3}(f_0+4f_1 +f_2)=\frac{h}{3}[f(a)+4f(a+h) +f(b)]
=\frac{b-a}{6}[a^2+4(\frac{a+b}{2})^2+b^2]=\frac{b^3}{3}-\frac{a^3}{3}\tag{2}$$
where $h=(b-a)/2$.<br>
<br>
For $f(x) = x^3$,
$$\int_{a}^{b} {f(x)} \,{\rm d}\approx \frac{h}{3}(f_0+4f_1 +f_2)=\frac{h}{3}[f(a)+4f(a+h) +f(b)]
=\frac{b-a}{6}[a^3+4(\frac{a+b}{2})^3+b^3]=\frac{b^4}{4}-\frac{a^4}{4}\tag{3}$$
